# CCCP Consulting Engine
## Consultants' Client Care Process — Interactive Workflow

**5 Phases:** Collect → Assess → Plan → Implement → Follow-Up

This notebook demonstrates the full CCCP pipeline using a simulated client engagement.
Each phase produces typed deliverables that feed the next — the same data flow
that powers the `nexcore-cccp` crate.

| Phase | What You Do | What You Get |
|:-----:|-------------|-------------|
| **P1** | Enter client context | System boundary map |
| **P2** | Review gap analysis | 15-dimensional proficiency gaps |
| **P3** | Prioritize interventions | Engagement plan with milestones |
| **P4** | Track progress | Completion dashboard |
| **P5** | Evaluate outcomes | Gap closure rate + disposition |

```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import json
from dataclasses import dataclass, field
from enum import IntEnum
from typing import Optional

# ── PV Domain Model (mirrors nexcore-cccp Rust types) ──

DOMAINS = [
    "D01: PV Foundations", "D02: Clinical Pharmacology", "D03: Regulatory Affairs",
    "D04: Case Management", "D05: Benefit-Risk Assessment", "D06: Signal Detection",
    "D07: Risk Management", "D08: Quality Systems", "D09: Data Management",
    "D10: Communication", "D11: Technology & AI", "D12: Special Populations",
    "D13: Global PV Operations", "D14: PV Leadership", "D15: Research & Innovation"
]

class Level(IntEnum):
    L1_NOVICE = 1
    L2_ADVANCED = 2
    L3_COMPETENT = 3
    L4_PROFICIENT = 4
    L5_EXPERT = 5

LEVEL_LABELS = {1: "L1 Novice", 2: "L2 Advanced Beginner", 3: "L3 Competent",
                4: "L4 Proficient", 5: "L5 Expert"}

# ── Engagement State ──

@dataclass
class EngagementState:
    client: str = ""
    org_size: str = "mid"
    markets: list = field(default_factory=lambda: ["US"])
    products: list = field(default_factory=lambda: ["small molecule"])
    current_levels: list = field(default_factory=lambda: [1]*15)
    desired_levels: list = field(default_factory=lambda: [3]*15)
    subsystems: dict = field(default_factory=dict)
    concerns: list = field(default_factory=list)
    interventions: list = field(default_factory=list)
    tracker: dict = field(default_factory=dict)
    final_levels: list = field(default_factory=lambda: [1]*15)

state = EngagementState()
print("✓ CCCP Engine initialized — 15 PV domains, 5 proficiency levels")
```

---
## Phase 1: Collect
*Algorithm: DEFINE + DECOMPOSE — Map the client's PV system boundaries*

Enter the client's organizational context and identify which PV subsystems exist, are partial, or are absent.

```python
# ── Phase 1: Client Intake ──

client_input = widgets.Text(value="Acme Pharmaceuticals", description="Client:")
size_input = widgets.Dropdown(options=["startup", "small", "mid", "large", "enterprise"],
                              value="mid", description="Org Size:")
markets_input = widgets.TagsInput(value=["US", "EU"], allowed_tags=["US", "EU", "Japan", "China", "ROW"],
                                  description="Markets:")
product_input = widgets.TagsInput(value=["small molecule"],
                                  allowed_tags=["small molecule", "biologic", "biosimilar", "device", "gene therapy"],
                                  description="Products:")

# Subsystem existence assessment
subsystem_names = ["Case Processing", "Signal Detection", "Risk Management",
                   "Regulatory Reporting", "Quality Oversight", "Data Management",
                   "ICSR Intake", "Benefit-Risk Assessment"]
subsystem_widgets = {}
for name in subsystem_names:
    subsystem_widgets[name] = widgets.ToggleButtons(
        options=["Exists", "Partial", "Absent"],
        value="Partial",
        description=f"{name}:",
        style={"description_width": "180px", "button_width": "80px"}
    )

concern_input = widgets.Textarea(
    value="Signal detection is entirely manual\nNo formal risk management process\nRegulatory deadlines occasionally missed",
    description="Concerns:", layout=widgets.Layout(width="600px", height="100px"),
    style={"description_width": "180px"}
)

p1_output = widgets.Output()

def save_p1(_):
    with p1_output:
        clear_output()
        state.client = client_input.value
        state.org_size = size_input.value
        state.markets = list(markets_input.value)
        state.products = list(product_input.value)
        state.subsystems = {name: w.value for name, w in subsystem_widgets.items()}
        state.concerns = [c.strip() for c in concern_input.value.split("\n") if c.strip()]

        gaps = [n for n, s in state.subsystems.items() if s == "Absent"]
        partial = [n for n, s in state.subsystems.items() if s == "Partial"]

        print(f"╔══ SYSTEM MAP: {state.client} ══╗")
        print(f"  Org: {state.org_size} | Markets: {', '.join(state.markets)} | Products: {', '.join(state.products)}")
        print(f"  Subsystems: {len(state.subsystems)} mapped")
        print(f"    ✓ Exists: {sum(1 for s in state.subsystems.values() if s == 'Exists')}")
        print(f"    ~ Partial: {len(partial)} — {', '.join(partial)}")
        print(f"    ✗ Absent: {len(gaps)} — {', '.join(gaps)}")
        print(f"  Concerns: {len(state.concerns)}")
        for c in state.concerns:
            print(f"    • {c}")
        print("╚══ Phase 1 Complete → Ready for Phase 2 ══╝")

save_btn = widgets.Button(description="Save System Map", button_style="warning")
save_btn.on_click(save_p1)

display(widgets.VBox([
    widgets.HTML("<h3>Organizational Context</h3>"),
    client_input, size_input, markets_input, product_input,
    widgets.HTML("<h3>Subsystem Existence Assessment</h3>"),
    *subsystem_widgets.values(),
    widgets.HTML("<h3>Presenting Concerns</h3>"),
    concern_input,
    save_btn, p1_output
]))
```

---
## Phase 2: Assess
*Algorithm: COMPARE + PERSPECTIVE — Quantify the 15-dimensional gap*

Set the current and desired proficiency levels for each domain. The gap analysis computes
the distance between current state and target state — `κ(current, desired) = |A△B|`.

```python
# ── Phase 2: Gap Analysis ──
import matplotlib.pyplot as plt
import numpy as np

current_sliders = []
desired_sliders = []

for i, domain in enumerate(DOMAINS):
    short = domain.split(": ")[1][:20]
    current_sliders.append(widgets.IntSlider(
        value=1, min=1, max=5, description=f"Current {short}:",
        style={"description_width": "200px"}, layout=widgets.Layout(width="450px")
    ))
    desired_sliders.append(widgets.IntSlider(
        value=3, min=1, max=5, description=f"Desired {short}:",
        style={"description_width": "200px"}, layout=widgets.Layout(width="450px")
    ))

p2_output = widgets.Output()

def run_gap_analysis(_):
    with p2_output:
        clear_output()
        state.current_levels = [s.value for s in current_sliders]
        state.desired_levels = [s.value for s in desired_sliders]
        gaps = [d - c for c, d in zip(state.current_levels, state.desired_levels)]

        # Radar chart
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), subplot_kw=dict(polar=True),
                                        facecolor="#0B0D13")
        short_names = [d.split(": ")[1][:12] for d in DOMAINS]
        angles = np.linspace(0, 2 * np.pi, 15, endpoint=False).tolist()
        angles += angles[:1]

        for ax, title, vals_c, vals_d in [
            (ax1, "Proficiency Levels", state.current_levels, state.desired_levels)
        ]:
            c_vals = vals_c + vals_c[:1]
            d_vals = vals_d + vals_d[:1]
            ax.set_facecolor("#141820")
            ax.plot(angles, c_vals, "o-", color="#DC2626", linewidth=2, label="Current")
            ax.fill(angles, c_vals, alpha=0.15, color="#DC2626")
            ax.plot(angles, d_vals, "o-", color="#D4AF37", linewidth=2, label="Desired")
            ax.fill(angles, d_vals, alpha=0.15, color="#D4AF37")
            ax.set_xticks(angles[:-1])
            ax.set_xticklabels(short_names, size=7, color="#7B95B5")
            ax.set_ylim(0, 5)
            ax.set_yticks([1,2,3,4,5])
            ax.set_yticklabels(["L1","L2","L3","L4","L5"], size=7, color="#7B95B5")
            ax.set_title("Current vs Desired", color="#D4AF37", pad=20)
            ax.legend(loc="upper right", fontsize=8)
            ax.tick_params(colors="#7B95B5")
            ax.spines["polar"].set_color("#D4AF37")
            ax.grid(color="#7B95B5", alpha=0.3)

        # Gap bar chart
        gap_vals = [max(0, g) for g in gaps]
        colors = ["#DC2626" if g >= 3 else "#D4AF37" if g >= 2 else "#7B95B5" for g in gap_vals]
        ax2.remove()
        ax2 = fig.add_subplot(122, facecolor="#141820")
        ax2.set_facecolor("#141820")
        bars = ax2.barh(short_names, gap_vals, color=colors)
        ax2.set_xlabel("Gap (levels)", color="#7B95B5")
        ax2.set_title("Priority Gaps", color="#D4AF37")
        ax2.tick_params(colors="#7B95B5")
        for spine in ax2.spines.values():
            spine.set_color("#7B95B5")
        ax2.invert_yaxis()

        plt.tight_layout()
        plt.show()

        # Summary stats
        total_gap = sum(max(0, g) for g in gaps)
        domains_met = sum(1 for g in gaps if g <= 0)
        readiness = domains_met / 15.0

        print(f"\n╔══ GAP ANALYSIS: {state.client} ══╗")
        print(f"  Domains meeting target: {domains_met}/15 ({readiness:.0%})")
        print(f"  Total gap distance: {total_gap} levels across {15 - domains_met} domains")
        if total_gap > 0:
            critical = [(DOMAINS[i], gaps[i]) for i in range(15) if gaps[i] >= 3]
            high = [(DOMAINS[i], gaps[i]) for i in range(15) if gaps[i] == 2]
            if critical:
                print(f"  Critical gaps (≥3): {', '.join(d.split(': ')[1] for d, _ in critical)}")
            if high:
                print(f"  High gaps (=2): {', '.join(d.split(': ')[1] for d, _ in high)}")
        print("╚══ Phase 2 Complete → Ready for Phase 3 ══╝")

p2_btn = widgets.Button(description="Run Gap Analysis", button_style="warning")
p2_btn.on_click(run_gap_analysis)

# Layout: current and desired side by side
rows = [widgets.HBox([current_sliders[i], desired_sliders[i]]) for i in range(15)]
display(widgets.VBox([*rows, p2_btn, p2_output]))
```

---
## Phase 3: Plan
*Algorithm: COMPOSE + ENERGY — Design the intervention sequence*

Automatically generates prioritized interventions from the gap analysis. Critical gaps (≥3 levels) become high-priority interventions. Edit priorities and session estimates as needed.

```python
# ── Phase 3: Engagement Plan ──

p3_output = widgets.Output()

def generate_plan(_):
    with p3_output:
        clear_output()
        gaps = [(i, state.desired_levels[i] - state.current_levels[i])
                for i in range(15) if state.desired_levels[i] > state.current_levels[i]]
        gaps.sort(key=lambda x: -x[1])  # largest gap first

        state.interventions = []
        for rank, (i, gap) in enumerate(gaps):
            domain = DOMAINS[i]
            priority = "CRITICAL" if gap >= 3 else "HIGH" if gap == 2 else "MEDIUM"
            sessions = min(gap + 1, 5)
            state.interventions.append({
                "id": f"INT-{rank+1:03d}",
                "domain": domain,
                "domain_idx": i,
                "gap": gap,
                "from": LEVEL_LABELS[state.current_levels[i]],
                "to": LEVEL_LABELS[state.desired_levels[i]],
                "priority": priority,
                "sessions": sessions,
            })

        total_sessions = sum(iv["sessions"] for iv in state.interventions)

        print(f"╔══ ENGAGEMENT PLAN: {state.client} ══╗")
        print(f"  Interventions: {len(state.interventions)} | Est. sessions: {total_sessions}")
        print(f"  {'─'*70}")
        print(f"  {'ID':<10} {'Domain':<25} {'Gap':>4} {'Priority':<10} {'Sessions':>8}")
        print(f"  {'─'*70}")
        for iv in state.interventions:
            short = iv["domain"].split(": ")[1][:24]
            print(f"  {iv['id']:<10} {short:<25} {iv['gap']:>4} {iv['priority']:<10} {iv['sessions']:>8}")
        print(f"  {'─'*70}")
        print(f"  {'TOTAL':<40} {'':<14} {total_sessions:>8}")
        print(f"\n  Sequence: {' → '.join(iv['id'] for iv in state.interventions[:5])}")
        if len(state.interventions) > 5:
            print(f"            ... + {len(state.interventions) - 5} more")
        print("╚══ Phase 3 Complete → Ready for Phase 4 ══╝")

p3_btn = widgets.Button(description="Generate Plan", button_style="warning")
p3_btn.on_click(generate_plan)
display(widgets.VBox([p3_btn, p3_output]))
```

---
## Phase 4: Implement
*Algorithm: COMPOSE + TRANSFER — Track execution progress*

Mark interventions as they progress. The tracker computes completion rate and surfaces blockers.

```python
# ── Phase 4: Implementation Tracker ──

p4_output = widgets.Output()
status_dropdowns = {}

def build_tracker(_=None):
    with p4_output:
        clear_output()
        if not state.interventions:
            print("⚠ Run Phase 3 first to generate the plan.")
            return

        status_dropdowns.clear()
        rows = []
        for iv in state.interventions:
            dd = widgets.Dropdown(
                options=["Not Started", "In Progress", "Blocked", "Completed", "Deferred"],
                value=state.tracker.get(iv["id"], "Not Started"),
                layout=widgets.Layout(width="140px")
            )
            status_dropdowns[iv["id"]] = dd
            short = iv["domain"].split(": ")[1][:20]
            label = widgets.HTML(f"<b>{iv['id']}</b> {short} (gap={iv['gap']})")
            rows.append(widgets.HBox([label, dd], layout=widgets.Layout(width="500px")))

        update_btn = widgets.Button(description="Update Tracker", button_style="warning")
        update_btn.on_click(update_tracker)
        display(widgets.VBox([*rows, update_btn]))

tracker_output = widgets.Output()

def update_tracker(_):
    with tracker_output:
        clear_output()
        state.tracker = {id: dd.value for id, dd in status_dropdowns.items()}

        total = len(state.tracker)
        completed = sum(1 for s in state.tracker.values() if s == "Completed")
        blocked = sum(1 for s in state.tracker.values() if s == "Blocked")
        in_progress = sum(1 for s in state.tracker.values() if s == "In Progress")
        rate = completed / max(total, 1)

        # Progress bar
        bar_width = 40
        filled = int(rate * bar_width)
        bar = "█" * filled + "░" * (bar_width - filled)

        print(f"╔══ IMPLEMENTATION: {state.client} ══╗")
        print(f"  [{bar}] {rate:.0%}")
        print(f"  Completed: {completed}/{total} | In Progress: {in_progress} | Blocked: {blocked}")
        if blocked > 0:
            blockers = [id for id, s in state.tracker.items() if s == "Blocked"]
            print(f"  ⚠ Blockers: {', '.join(blockers)}")
        print("╚══ Update status above and click Update Tracker ══╝")

init_btn = widgets.Button(description="Initialize Tracker", button_style="info")
init_btn.on_click(build_tracker)
display(widgets.VBox([init_btn, p4_output, tracker_output]))
```

---
## Phase 5: Follow-Up
*Algorithm: COMPARE + PERSPECTIVE — Measure gap closure and determine disposition*

Set the client's final proficiency levels after interventions. The evaluation computes
gap closure rate and recommends a disposition (Close, Continue, Transition, or Pause).

```python
# ── Phase 5: Outcome Evaluation ──

final_sliders = []
for i, domain in enumerate(DOMAINS):
    short = domain.split(": ")[1][:20]
    final_sliders.append(widgets.IntSlider(
        value=3, min=1, max=5, description=f"Final {short}:",
        style={"description_width": "200px"}, layout=widgets.Layout(width="500px")
    ))

p5_output = widgets.Output()

def evaluate_outcomes(_):
    with p5_output:
        clear_output()
        state.final_levels = [s.value for s in final_sliders]

        initial_gaps = [max(0, d - c) for c, d in zip(state.current_levels, state.desired_levels)]
        final_gaps = [max(0, d - f) for f, d in zip(state.final_levels, state.desired_levels)]

        initial_total = sum(initial_gaps)
        final_total = sum(final_gaps)
        closure_rate = 1.0 - (final_total / max(initial_total, 1))

        # Disposition logic
        if closure_rate >= 0.8:
            disposition = "CLOSE — Objectives achieved, client self-sustaining"
            disp_color = "#22c55e"
        elif closure_rate >= 0.5:
            disposition = "CONTINUE — Significant progress, more work needed"
            disp_color = "#D4AF37"
        elif closure_rate >= 0.2:
            disposition = "TRANSITION — Scope change required for remaining gaps"
            disp_color = "#f97316"
        else:
            disposition = "PAUSE — Limited progress, investigate blockers"
            disp_color = "#DC2626"

        # Before/After comparison chart
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), facecolor="#0B0D13")
        short_names = [d.split(": ")[1][:12] for d in DOMAINS]
        x = np.arange(15)

        # Proficiency comparison
        ax1.set_facecolor("#141820")
        ax1.bar(x - 0.2, state.current_levels, 0.2, label="Initial", color="#DC2626", alpha=0.8)
        ax1.bar(x, state.final_levels, 0.2, label="Final", color="#22c55e", alpha=0.8)
        ax1.bar(x + 0.2, state.desired_levels, 0.2, label="Target", color="#D4AF37", alpha=0.5)
        ax1.set_xticks(x)
        ax1.set_xticklabels(short_names, rotation=45, ha="right", fontsize=7, color="#7B95B5")
        ax1.set_ylabel("Proficiency Level", color="#7B95B5")
        ax1.set_title("Initial → Final → Target", color="#D4AF37")
        ax1.legend(fontsize=8)
        ax1.set_ylim(0, 5.5)
        ax1.tick_params(colors="#7B95B5")
        for spine in ax1.spines.values():
            spine.set_color("#7B95B5")

        # Gap closure waterfall
        ax2.set_facecolor("#141820")
        ax2.barh(short_names, initial_gaps, color="#DC2626", alpha=0.4, label="Initial Gap")
        ax2.barh(short_names, final_gaps, color="#DC2626", alpha=0.9, label="Remaining Gap")
        ax2.set_xlabel("Gap (levels)", color="#7B95B5")
        ax2.set_title(f"Gap Closure: {closure_rate:.0%}", color="#D4AF37")
        ax2.legend(fontsize=8)
        ax2.tick_params(colors="#7B95B5")
        for spine in ax2.spines.values():
            spine.set_color("#7B95B5")
        ax2.invert_yaxis()

        plt.tight_layout()
        plt.show()

        # Residual gaps
        residual = [DOMAINS[i].split(": ")[1] for i in range(15) if final_gaps[i] > 0]

        print(f"\n╔══ OUTCOME EVALUATION: {state.client} ══╗")
        print(f"  Gap Closure: {closure_rate:.0%} ({initial_total} → {final_total} total gap levels)")
        print(f"  Disposition: {disposition}")
        if residual:
            print(f"  Residual Gaps: {', '.join(residual)}")
        else:
            print(f"  All domains meet target!")
        print("╚══ CCCP Engagement Complete ══╝")

p5_btn = widgets.Button(description="Evaluate Outcomes", button_style="success")
p5_btn.on_click(evaluate_outcomes)
display(widgets.VBox([*final_sliders, p5_btn, p5_output]))
```
